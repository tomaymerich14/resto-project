```python
baseline = pd.DataFrame()
```

```python
baseline['jour']=['Lundi','Mardi','Mercredi','Jeudi','Vendredi','Samedi','Dimanche']
```

```python
baseline = pd.DataFrame(np.repeat(baseline.values,2,axis=0),columns=['jour'])
```

```python
baseline['service']='soir'
```

```python
for i in np.linspace(0,12,7):
    baseline['service'][i]='midi'
```

```python
baseline
```

```python
j = ['Lundi','Mardi','Mercredi','Jeudi','Vendredi','Samedi','Dimanche']
for jour in j:
    print(jour)
```

```python
ser = ['midi','soir']
```

```python
baseline_d2 = baseline
```

```python
baseline_d16 = baseline
```

```python
baseline_d16
```

```python
score_d2 = []

for i in j:  
    print(i)
    for s in ser:
        print(s)
        score_d2.append(df_d2[(df_d2.jour == i) & (df_d2.service == s)]['CA_TTC'].mean())
```

```python
score_d16 = []

for i in j:  
    print(i)
    for s in ser:
        print(s)
        score_d16.append(df_d16[(df_d16.jour == i) & (df_d16.service == s)]['CA_TTC'].mean())
```

## D16 

```python
baseline_d16['moyenne'] =score_d16
```

```python
baseline_d16
```

```python
df_d16['baseline_pred']=''
```

```python
for i in range(len(df_d16)):  
    jour = df_d16['jour'][i] 
    service = df_d16['service'][i] 
    base_score = baseline_d16[(baseline_d16.jour == jour) & (baseline_d16.service == service)][['moyenne']].iloc[0][0]
    df_d16['baseline_pred'][i]=base_score 
```

```python
df_d16
```

```python
df_d16['difference']=abs(df_d16['CA_TTC']-df_d16['baseline_pred'])
```

```python
df_d16['difference'].mean()
```
